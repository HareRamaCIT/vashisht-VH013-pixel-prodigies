```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Average
# Define the parameters
num_values = 1000  # Number of values
mean = 60          # Mean of the normal distribution
std_dev = 15       # Standard deviation of the normal distribution
min_value = 0      # Minimum value
max_value = 100    # Maximum value
random_seed = 42

# Generate random data following a normal distribution
np.random.seed(random_seed)
data = np.random.normal(loc=mean, scale=std_dev, size=num_values)

# Ensure values are within the desired range
data = np.clip(data, min_value, max_value)
# Round the values to the nearest integer
data = np.round(data).astype(int)
np.random.shuffle(data)
# Create a pandas DataFrame
df1 = pd.DataFrame(data, columns=['avg_score'])

# Plot a histogram of the "values" column
plt.hist(df1['avg_score'], bins=20, alpha=0.7)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Average Score')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Display the first few rows of the DataFrame
df1.describe()
```

```python
# Set the parameters
num_values = 1000  # Number of values to generate
prob_success = 0.75  # Probability of success (passing the quiz on an attempt)

# Set the random seed for reproducibility
np.random.seed(random_seed)

# Generate random data following a geometric distribution
data = np.random.geometric(p=prob_success, size=num_values)
# Clip values to be within the desired range (0 to 8)
data = np.clip(data, 0, 8)
np.random.shuffle(data)
# Convert the data to a pandas DataFrame
df2 = pd.DataFrame(data, columns=['avg_attempts'])
# Plot a histogram of the generated data using pandas plotting
df2['avg_attempts'].plot(kind='hist', bins=np.arange(0, 9)-0.5, density=True, alpha=0.7)
plt.xlabel('Number of Attempts')
plt.ylabel('Probability Density')
plt.title('Average Number of Attempts')
plt.xticks(range(9))
plt.grid(axis='y', alpha=0.75)
plt.show()
df2.describe()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters
num_values = 1000  # Number of values to generate
mean = 5           # Mean of the truncated normal distribution
std_dev = 3        # Standard deviation of the truncated normal distribution
min_value = 0      # Minimum value
max_value = 16     # Maximum value

# Set the random seed for reproducibility
np.random.seed(random_seed)

# Generate random data following a truncated normal distribution
data = np.random.normal(loc=mean, scale=std_dev, size=num_values)
data = np.clip(data, min_value, max_value)  # Clip values within the desired range
np.random.shuffle(data)
# Convert the data to a pandas DataFrame
df3 = pd.DataFrame(data, columns=['avg_learn_time'])

# Plot a histogram of the generated data using pandas plotting
df3['avg_learn_time'].plot(kind='hist', bins=30, density=True, alpha=0.7)
plt.xlabel('Average Learn Time (hours)')
plt.ylabel('Probability Density')
plt.title('Average Learn Time')
plt.grid(axis='y', alpha=0.75)
plt.show()
df3.describe()
```

```python
# Set the parameters
num_values = 1000  # Number of values to generate
alpha = 4  # Shape parameter for the beta distribution
beta = 4   # Shape parameter for the beta distribution
random_seed = 42  # Random seed for reproducibility

# Set the random seed for reproducibility
np.random.seed(random_seed)

# Generate random data following a beta distribution
data = np.random.beta(alpha, beta, size=num_values)
np.random.shuffle(data)

# Convert the data to a pandas DataFrame
df5 = pd.DataFrame(data, columns=['matrl_compln'])

# Plot a histogram of the generated data
plt.hist(df5, bins=30, alpha=0.7)
plt.xlabel('Material Completion Rate')
plt.ylabel('Frequency')
plt.title('Distribution of Material Completion Rates')
plt.grid(axis='y', alpha=0.75)
plt.show()

df5.describe()
```

```python
df = pd.concat([df1, df2, df3, df4, df5, df6], axis=1)
df
```

```python
y_predicted = km.fit_predict(df)
y_predicted
```

```python
df['learner_category'] = y_predicted
df['learner_category'].value_counts()
```

```python
df1 = df[df["learner_category"]==0]
df2 = df[df["learner_category"]==1]
df3 = df[df["learner_category"]==2]

plt.scatter(df1["avg_score"], df1["learner_category"],  color='green')
plt.scatter(df2["avg_score"], df2["learner_category"],  color='red')
plt.scatter(df3["avg_score"], df3["learner_category"],  color='blue')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Assuming df contains your dataset with multiple features including "learner_category"

# Separate features and target variable
X = df.drop(columns=['learner_category'])  # Features
y = df['learner_category']  # Target variable

# Standardize the features (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce dimensions
pca = PCA(n_components=2)  # Reduce to 2 components for plotting
X_pca = pca.fit_transform(X_scaled)

# Fit K-means clustering algorithm
kmeans = KMeans(n_clusters=3, random_state=42)  # Assuming 3 clusters
kmeans.fit(X_scaled)
cluster_labels = kmeans.labels_

# Create DataFrame with reduced dimensions and cluster labels
df_clustered = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
df_clustered['Cluster'] = cluster_labels

# Plot clusters based on reduced dimensions
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df_clustered['PC1'], df_clustered['PC2'], c=df_clustered['Cluster'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Clusters based on PCA')
plt.legend(*scatter.legend_elements(), title='Clusters')
plt.show()
```
